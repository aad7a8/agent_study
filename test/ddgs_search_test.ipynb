# DDGS Search Function Test

使用 `ddgs` 套件測試各種搜尋功能（無需 API key）

```python
from ddgs import DDGS
```

## 1. 文字搜尋

```python
def text_search(query: str, max_results: int = 10, region: str = "tw-tzh"):
    """
    文字搜尋
    :param query: 搜尋關鍵字
    :param max_results: 最多回傳幾筆結果
    :param region: 搜尋地區，預設台灣繁體中文
    """
    results = DDGS().text(query, region=region, max_results=max_results)
    for i, r in enumerate(results, 1):
        print(f"[{i}] {r['title']}")
        print(f"    URL : {r['href']}")
        print(f"    摘要 : {r['body'][:100]}...")
        print()
    return results

results = text_search("Python 教學")
```

## 2. 新聞搜尋

```python
def news_search(query: str, max_results: int = 5, timelimit: str = "w"):
    """
    新聞搜尋
    :param query: 搜尋關鍵字
    :param max_results: 最多回傳幾筆結果
    :param timelimit: 時間範圍 d=天, w=週, m=月
    """
    results = DDGS().news(query, max_results=max_results, timelimit=timelimit)
    for i, r in enumerate(results, 1):
        print(f"[{i}] {r['title']}")
        print(f"    來源 : {r['source']}")
        print(f"    時間 : {r['date']}")
        print(f"    URL  : {r['url']}")
        print()
    return results

results = news_search("AI 人工智慧")
```

## 3. 圖片搜尋

```python
def image_search(query: str, max_results: int = 5):
    """
    圖片搜尋
    :param query: 搜尋關鍵字
    :param max_results: 最多回傳幾筆結果
    """
    results = DDGS().images(query, max_results=max_results)
    for i, r in enumerate(results, 1):
        print(f"[{i}] {r['title']}")
        print(f"    圖片 URL : {r['image']}")
        print(f"    來源頁面 : {r['url']}")
        print()
    return results

results = image_search("台灣風景")
```

## 4. 影片搜尋

```python
def video_search(query: str, max_results: int = 5):
    """
    影片搜尋
    :param query: 搜尋關鍵字
    :param max_results: 最多回傳幾筆結果
    """
    results = DDGS().videos(query, max_results=max_results)
    for i, r in enumerate(results, 1):
        print(f"[{i}] {r['title']}")
        print(f"    時長     : {r.get('duration', 'N/A')}")
        print(f"    發布者   : {r.get('publisher', 'N/A')}")
        print(f"    URL      : {r['content']}")
        print()
    return results

results = video_search("Python tutorial")
```

## 5. 整合搜尋（一次搜尋多種類型）

```python
def unified_search(query: str, max_results: int = 3):
    """
    整合搜尋：同時搜尋文字、新聞、圖片
    """
    print(f"{'='*50}")
    print(f"搜尋關鍵字：{query}")
    print(f"{'='*50}\n")

    print(">>> 文字結果")
    text_search(query, max_results=max_results)

    print(">>> 新聞結果")
    news_search(query, max_results=max_results)

    print(">>> 圖片結果")
    image_search(query, max_results=max_results)

unified_search("Claude AI")
```

# DDGS Dorking

利用搜尋引擎的進階語法（Search Operators）精確縮小搜尋範圍。

| Operator | 說明 | 範例 |
|----------|------|------|
| `site:` | 限定網域 | `site:github.com` |
| `filetype:` | 限定檔案類型 | `filetype:pdf` |
| `inurl:` | URL 包含關鍵字 | `inurl:admin` |
| `intitle:` | 標題包含關鍵字 | `intitle:login` |
| `intext:` | 內文包含關鍵字 | `intext:password` |
| `"phrase"` | 精確詞組 | `"exact phrase"` |
| `-word` | 排除關鍵字 | `-ads` |
| `OR` | 或條件 | `python OR javascript` |

> **注意**：各 backend 對 operator 的支援程度不同，建議使用 `backend="google"` 或 `backend="bing"` 以獲得最佳相容性。

```python
def dork_search(dork_query: str, max_results: int = 5, backend: str = "auto", timelimit: str | None = None):
    """
    Dorking 搜尋：支援進階搜尋語法
    :param dork_query: 含有 operator 的搜尋字串，例如 'site:github.com filetype:pdf python'
    :param max_results: 最多回傳幾筆結果
    :param backend: 搜尋引擎 (auto/google/bing/duckduckgo/brave/mojeek/yandex/yahoo/wikipedia)
    :param timelimit: 時間範圍 d=天 / w=週 / m=月 / y=年
    """
    print(f"Dork Query : {dork_query}")
    print(f"Backend    : {backend}")
    print("-" * 60)
    results = DDGS().text(
        dork_query,
        safesearch="off",
        timelimit=timelimit,
        max_results=max_results,
        backend=backend,
    )
    for i, r in enumerate(results, 1):
        print(f"[{i}] {r['title']}")
        print(f"    URL  : {r['href']}")
        print(f"    摘要 : {r['body'][:120]}...")
        print()
    return results
```

## 6. `filetype:` — 搜尋特定檔案類型

```python
# 搜尋 PDF 格式的 Python Machine Learning 資源
dork_search("python machine learning filetype:pdf")
```

```python
# 搜尋 CSV 資料集
dork_search("dataset filetype:csv climate change")
```

## 7. `site:` — 限定網域搜尋

```python
# 只搜尋 GitHub 上的結果
dork_search("site:github.com python web scraping")
```

```python
# 搜尋 Reddit 上的討論（近一個月）
dork_search("site:reddit.com best Python library", timelimit="m")
```

## 8. `intitle:` / `inurl:` — 標題與 URL 條件

```python
# 標題含有 "tutorial" 的 FastAPI 頁面
dork_search("intitle:tutorial python fastapi")
```

```python
# URL 包含 "api" 的文件頁面
dork_search("inurl:api docs python")
```

## 9. 組合 Dork — 多個 Operator 疊加

```python
# site: + filetype: 組合：GitHub 上的 PDF 資源
dork_search("site:github.com filetype:pdf python data science")
```

```python
# 精確詞組 + 排除關鍵字 + site: + 時間限制
dork_search('"machine learning" -ads site:medium.com', timelimit="m")
```

```python
# OR 條件：搜尋 Python 或 JavaScript 的教學
dork_search("intitle:tutorial python OR javascript site:dev.to")
```

## 10. 指定 Backend 的 Dorking

```python
# Google backend — operator 相容性最佳
dork_search("site:stackoverflow.com intitle:python asyncio", backend="google")
```

```python
# Bing backend
dork_search("filetype:pdf deep learning", backend="bing")
```
